```python
#First graph: contribution moyenne par décile:
df_rev=df.loc[~df["Count_pourmil_CP"].isnull()]
df_rev=df_rev.loc[~df_rev["perc_ecolo"].isnull()]
df_rev=df_rev.loc[~df_rev["perc_fisca"].isnull()]
df_rev=df_rev.loc[~df_rev["perc_etat"].isnull()]
df_rev=df_rev.loc[~df_rev["perc_cit"].isnull()]

contrib_mean_rev=list()
contrib_fisca=list()
contrib_ecolo=list()
contrib_etat=list()
contrib_cit=list()
quantile=np.linspace(0,0.90,10)
rev_median=list()

for i in quantile:
    pop_quantile=sum(df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"POP_CP"])
    contrib_mean_rev.append(sum(df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"Count_pourmil_CP"]*
                           df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"POP_CP"])/pop_quantile)
    contrib_fisca.append(sum(df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"perc_fisca"]*
                           df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"POP_CP"])/pop_quantile)
    contrib_ecolo.append(sum(df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"perc_ecolo"]*
                           df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"POP_CP"])/pop_quantile)
    contrib_etat.append(sum(df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"perc_etat"]*
                           df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"POP_CP"])/pop_quantile)
    contrib_cit.append(sum(df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"perc_cit"]*
                           df_rev.loc[(df_rev["C15_POP15P_CS3mean_CP"]>df_rev["C15_POP15P_CS3mean_CP"].quantile(i))&(df_rev["C15_POP15P_CS3mean_CP"]<df_rev["C15_POP15P_CS3mean_CP"].quantile((i+0.05))),"POP_CP"])/pop_quantile)
    
    rev_median.append(df_rev["C15_POP15P_CS3mean_CP"].quantile(i+0.05))
    
    
print(contrib_mean_rev) 
print(contrib_fisca) 

plot_data = pd.DataFrame(
    {'Mean_Contrib': contrib_mean_rev,
     'quantile': quantile,
     'rev_median': rev_median,
     'perc_fisca':contrib_fisca,
     'perc_ecolo':contrib_ecolo,
     'perc_etat':contrib_etat,
     'perc_cit':contrib_cit,
    })

ax=sns.lmplot(x="rev_median", y="Mean_Contrib", data=plot_data, fit_reg=False)
#ax = sns.barplot(x="quantile", y="Mean_Contrib", data=plot_data,  palette="Blues_d")


```

```python
barWidth = 0.1
plt.bar(plot_data["quantile"].tolist() ,plot_data["perc_fisca"].tolist() , color='#7f6d5f', edgecolor='white', width=barWidth)
# Create green bars (middle), on top of the firs ones
plt.bar(plot_data["quantile"].tolist() ,plot_data["perc_ecolo"].tolist() ,bottom=plot_data["perc_fisca"].tolist(), color='#557f2d', edgecolor='white', width=barWidth)
# Create green bars (top)
plt.bar(plot_data["quantile"].tolist() ,plot_data["perc_etat"].tolist() ,bottom=(plot_data["perc_fisca"]+plot_data["perc_ecolo"]).tolist(), color='#557f2d', edgecolor='white', width=barWidth)
plt.bar(plot_data["quantile"].tolist() ,plot_data["perc_cit"].tolist() ,bottom=(plot_data["perc_fisca"]+plot_data["perc_ecolo"]+plot_data["perc_etat"]).tolist(), color='#557f2d', edgecolor='white', width=barWidth)

plt.show()
```
